# Method of removing outliers from the visualised boxplot:

Sometimes a dataset can contain extreme values that are outside the range of what is expected and unlike the other data. These are called outliers and often machine learning modeling and model skill in general can be improved by understanding and even removing these outlier values.


This does not mean that the values identified are outliers and should be removed.

For example, If we consider the column "Annual_Premium", there are n number of outliers. But, if we observe the column ANNUAL PREMIUM refers to "The Premium paid by a customer annually". If a person X paid the highest annual premium is considered to be an outlier and removed, The mean value of the column gets affected. 



So, Removing Outliers all the time is not a good method.

### Spotting the outliers using Z-Score method:

```python
from scipy import stats
import numpy as np
 
z = np.abs(stats.zscore(df['Driving_License']))
print(z)
```

### Spotting the Outliers using IQR method:

```python
Q1 = np.percentile(df['Driving_License'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['Driving_License'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
# Above Upper bound
upper = df['Driving_License'] >= (Q3+1.5*IQR)
 
print("Upper bound:",upper)
print(np.where(upper))
 
# Below Lower bound
lower = df['Driving_License'] <= (Q1-1.5*IQR)
print("Lower bound:", lower)
print(np.where(lower))
```

# Removing the Outliers after the whole analysis.

```python
Q1 = np.percentile(df['Annual_Premium'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['Annual_Premium'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", df.shape)
 
# Upper bound
upper = np.where(df['Annual_Premium'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['Annual_Premium'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
 
print("New Shape: ", df.shape)
```

# Multivariate Analysis

Multivariate analysis is based in observation and analysis of more than one statistical outcome variable at a time. In design and analysis, the technique is used to perform trade studies across multiple dimensions while taking into account the effects of all variables on the responses of interest. The development of multivariate methods emerged to analyze large databases and increasingly complex data. 


Since the best way to represent the knowledge of reality is the modeling, we should use multivariate statistical methods. Multivariate methods are designed to simultaneously analyze data sets, i.e., the analysis of different variables for each person or object studied. Keep in mind at all times that all variables must be treated accurately reflect the reality of the problem addressed. There are different types of multivariate analysis and each one should be employed according to the type of variables to analyze: dependent, interdependence and structural methods. In conclusion, multivariate methods are ideal for the analysis of large data sets and to find the cause and effect relationships between variables; there is a wide range of analysis types that we can use.

Multivariate analysis in a broad sense is the set of statistical methods aimed simultaneously analyze datasets. That is, for each individual or object being studied, analyzed several variables. The essence of multivariate thinking is to expose the inherent structure and meaning revealed within these sets if variables through application and interpretation of various statistical methods.
